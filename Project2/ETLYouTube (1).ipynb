```python
US_data = "USvideos.csv"
UK_data = "GBvideos.csv"
CA_data = "CAvideos.csv"
```

```python
us_videos = pd.read_csv(US_data)
uk_videos = pd.read_csv(UK_data)
ca_videos = pd.read_csv(CA_data)
```

```python
dropped_us = us_videos.drop(['trending_date','comments_disabled',
                             'ratings_disabled','video_error_or_removed', 'thumbnail_link'], axis = 1)
dropped_uk = uk_videos.drop(['trending_date','comments_disabled',
                             'ratings_disabled','video_error_or_removed', 'thumbnail_link'], axis = 1)
dropped_ca = ca_videos.drop(['trending_date','comments_disabled',
                             'ratings_disabled','video_error_or_removed', 'thumbnail_link'], axis = 1)
```

```python
dropped_us.head()
```

```python
#possible ways to join data:
#? Maybe get a list of categories and match it to id
#  Possible match video_id common in all three tables, get a combined csv file that has all videos
#  Get a list of category numbers, and get translation from category_id to category name
#  If we combine the data, we may need to use sqlite or nosql because the data would be relational
#  group by category, maybe group by channel, 

```

```python
category_list = open("categories.txt", "r").read()
category_list = category_list.split('\n')
category_translation = {}
for category in category_list:
    number = category.split(" ")[0]
    name = ''.join(category.split(" ")[1:])
    category_translation[int(number)] = name
```

```python
category_translation
```

```python
dropped_us = dropped_us.replace({'category_id': category_translation })
dropped_uk = dropped_uk.replace({'category_id': category_translation })
dropped_ca = dropped_ca.replace({'category_id': category_translation })
```

```python
dropped_us
```

```python
combined_videos = pd.merge(dropped_us, dropped_uk, how = 'inner', on = 'title')
combined_videos = pd.merge(combined_videos, dropped_ca, how = 'inner', on = 'title')
combined_videos.columns
#print(combined_videos["title"].nunique())
#combined_videos.columns
combined_videos = combined_videos[['channel_title_x', 'category_id_x', 'tags_x']]
combined_videos = combined_videos.drop_duplicates()
combined_videos = combined_videos.reset_index(drop = True)
```

```python
combined_videos
```

```python
top_500 = "top_500.csv"
```

```python
top_500_df = pd.read_csv(top_500)
top_500_df
```

```python
combined_channels = pd.merge(top_500_df, combined_videos, how = 'inner', left_on = 'Ch_name', right_on = 'channel_title_x')
```

```python
combined_channels = combined_channels[['Rank','Grade','Ch_name','Uploads','Subscriptions','Views','category_id_x']]
combined_channels = combined_channels.drop_duplicates()
combined_channels = combined_channels.reset_index(drop = True)
```

```python
combined_channels_dict  = combined_channels.to_dict()
channel_ranks = combined_channels_dict["Rank"]
for index, rank in channel_ranks.items():
    rank_integer = int(rank.strip()[:-2])
    combined_channels_dict["Rank"][index] = rank_integer
```

```python
channel_ranks
```

```python
Channel_Ranks = pd.DataFrame.from_dict(combined_channels_dict)
Channel_Ranks = Channel_Ranks.sort_values(by='Rank', ascending = True)
Channel_Ranks.rename(columns = {'Ch_name':'Channel','category_id_x':'Category'}, inplace = True)
Channel_Ranks
```
