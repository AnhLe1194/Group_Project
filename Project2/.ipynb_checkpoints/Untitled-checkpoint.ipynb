```python
import pandas as pd
```

```python
CA_df = pd.read_csv("CAvideos.csv")
GB_df = pd.read_csv("GBvideos.csv")
US_df = pd.read_csv("USvideos.csv")
```

```python
CA_df.head()
```

```python
dropped_ca = CA_df.drop(['trending_date','publish_time','comments_disabled','ratings_disabled','video_error_or_removed', 'thumbnail_link'], axis = 1)
dropped_ca
```

```python
dropped_us = US_df.drop(['trending_date','publish_time','comments_disabled','ratings_disabled','video_error_or_removed', 'thumbnail_link'], axis = 1)
dropped_us
```

```python
dropped_gb = GB_df.drop(['trending_date','publish_time','comments_disabled','ratings_disabled','video_error_or_removed', 'thumbnail_link'], axis = 1)
dropped_gb
```

```python
print(US_df.shape)
#rows, columns
```

```python
category_list = open("categories.txt", "r").read()
category_list = category_list.split('\n')
category_translation = {}
for category in category_list:
    number = category.split(" ")[0]
    name = ''.join(category.split(" ")[1:])
    category_translation[int(number)] = name
```

```python
category_translation
```

```python
dropped_us = dropped_us.replace({'category_id': category_translation})
dropped_gb = dropped_gb.replace({'category_id': category_translation})
dropped_ca = dropped_ca.replace({'category_id': category_translation})
```

```python
dropped_us.head()
```

```python
#dropped_us.merge(dropped_gb, how = 'inner', on = "video_id")
combined_videos = pd.merge(dropped_us, dropped_gb, how = 'inner', on = "video_id")
combined_videos = pd.merge(combined_videos, dropped_ca, how = 'inner', on = "video_id")
combined_videos.columns
combined_videos = combined_videos[['title_x', 'channel_title_x', 'category_id_x', 'tags_x']]
combined_videos = combined_videos.drop_duplicates()
combined_videos = combined_videos.reset_index(drop = True)
```
